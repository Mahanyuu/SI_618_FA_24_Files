- üñºÔ∏è **Creates a figure** to hold your plots.

- üìä **Plots some data**, such as lines or scatter points.
- üé® **Decorates the plot** with labels, titles, and other features to enhance its readability and appearance.

````python
# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
````

To create a **figure** with just **one subplot**, you can use the following:


````python
f, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
plt.show()
````

To create a **figure** with **two subplots** in a 1-D array of axes:


````python
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(x, y)
axarr[0].set_title('Sharing X axis')
axarr[1].scatter(x, y)
plt.show()
````

Two subplots, unpack the axes array immediately


````python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
````

To create a **figure** with **three subplots** sharing both **x** and **y** axes:


````python
f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing both axes')
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')
plt.show()
````

Fine-tune figure; make subplots close to each other and hide x ticks for all but bottom plot.

````python
f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing both axes')
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')

f.subplots_adjust(hspace=0)
plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
plt.show()
````

Row and column sharing:

````python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
ax1.plot(x, y)
ax1.set_title('Sharing x per column, y per row')
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')
ax4.plot(x, 2 * y ** 2 - 1, color='r')
plt.show()
````

## üìå **Create a Stacked Bar Plot** üìä

## üìå  Subplots (again)

In addition to the way we used **subplots** earlier in this class, we can use the **`.subplots()`** function to generate **multiple plots** within a figure. The **`.subplots()`** function returns a set of axes on which we can make plots.

To demonstrate how this works, let's start by filling in just one of the subplots:


````python
f, (ax1, ax2) = plt.subplots(2)  # if only 1 argument, we assume it's the number of rows
ax1.hist(menu['Calories'])
plt.show()
````

Now let's fill in both subplots:

````python
f, (ax1, ax2) = plt.subplots(2)
ax1.hist(menu['Calories'])
ax2.plot(menu['Calories'], menu['Total Fat'], 'bo')
plt.show()
````

Now let's make a **2x2 layout** of 4 plots. üñºÔ∏è  
Note the structure of the return values from the **`.subplots()`** function: it returns a **2x2 array of axes**. Each subplot can be accessed using **`axes[row, col]`**.

This layout allows us to organize multiple plots in a single figure, making comparisons easier and more visually structured.

````python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.hist(menu['Calories'])
ax2.plot(menu['Calories'], menu['Total Fat'], 'bo')
plt.show()
````

````python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.hist(menu['Total Fat'])
ax2.plot(menu['Calories'], menu['Total Fat'], 'bo')
ax3.hist(menu['Dietary Fiber'])
ax4.plot(menu['Calories'], menu['Total Fat'], 'bo')
plt.show()
````

Alternatively, we can use the **Pandas-Matplotlib integration** for plotting.  
Note the use of the **`ax=`** keyword argument, which allows you to specify the axes where the plot should appear within a subplot layout.

This makes it easy to generate multiple plots in different subplots while working directly with Pandas data structures.

````python
f, (ax1, ax2) = plt.subplots(2)
menu['Calories'].plot(ax=ax1, kind='hist')
ax2.plot()
plt.show()
````

## üìå **Why Do We Care About Distributions?** üìä

# üåü Poisson

$$ P(k~events~in~interval) = \frac{\lambda^ke^{-\lambda}}{k!} $$

$\lambda$ represents the **event rate** (i.e., the average number of occurrences in a given time period).

### **Examples**:
- üå† **Meteor strikes**
- üè• **Arrival of patients** to a hospital

````python
# as lambda goes up --> looks more normal
pois = np.random.poisson(3, 100000) # lambda, count 
sns.displot(pois)
````

# üåü **Power/Zipf/Pareto Distributions**

The general formula for these distributions is:

$$ P = \frac{x^{-a}}{\zeta(a)} $$

### **Characteristics of "long tail" distributions**:
- **Degree distribution** (e.g., in networks)
- **Movie/music popularity** (e.g., a few hits, many less popular items)
- **Words** (e.g., common vs. rare words in a language)

````python
power = np.random.zipf(2, 100000)
sns.displot(power, log_scale=(True, True))
````

## üìå **Run Sequence** üìà

- A **Run Sequence** plot shows **index versus value**.
- If the plot is **flat** and **non-drifting**:
  - The **fixed-location assumption** holds.
- If the **vertical spread** is consistent across the entire plot:
  - The **fixed-variation assumption** holds.

````python
ax = sns.regplot(x=np.arange(len(testdata)), y=testdata, fit_reg=False)
ax.set_ylim(0, 250)
ax.set_ylabel("val")
ax.set_xlabel("i")
````

# üåü **Lag Plot** üìä

- Plot point $y_i$ versus $y_{i-1}$.
- If the **lag plot** is **structureless**:
  - The **randomness assumption** holds.

````python
lag = testdata.copy()
lag = np.array(lag[:-1])
current = testdata[1:]
ax = sns.regplot(x=current, y=lag, fit_reg=False)
ax.set_ylabel("y_i-1")
ax.set_xlabel("y_i")
````

## üìå **QQ Plot** üìà

- A **QQ Plot** takes our **n ordered data points**:
  - Sorted from smallest to largest.
- It asks:
  - What is the relationship between **quantiles from our data** and **quantiles from a theoretical distribution** that we're assuming the sample is drawn from?

````python
from scipy import stats
````

````python
qntls, xr = stats.probplot(testdata, fit=False)
sns.regplot(y=xr, x=qntls)
````

Another way to do exactly the same thing:

````python
import statsmodels.api as sm

sm.qqplot(testdata)
plt.show()
````

````python
def random_snorm(n, mean = 0, sd = 1, xi = 1.5):
    def random_snorm_aux(n, xi):
        weight = xi/(xi + 1/xi)
        z = np.random.uniform(-weight,1-weight,n)
        xi_ = xi**np.sign(z)
        random = -np.absolute(np.random.normal(0,1,n))/xi_ * np.sign(z)
        m1 = 2/np.sqrt(2 * np.pi)
        mu = m1 * (xi - 1/xi)
        sigma = np.sqrt((1 - m1**2) * (xi**2 + 1/xi**2) + 2 * m1**2 - 1)
        return (random - mu)/sigma

    return random_snorm_aux(n, xi) * sd + mean
````

````python
rightskewed = random_snorm(1000, xi=2)*100
sns.histplot(rightskewed, kde=False)
````

````python
qntls, xr = stats.probplot(rightskewed, fit=False)
sns.regplot(x=xr, y=qntls)
````

````python
leftskewed = random_snorm(1000, xi=-2)*100
sns.histplot(leftskewed, kde=False)
````

````python
qntls, xr = stats.probplot(leftskewed, fit=False)
sns.regplot(x=xr, y=qntls)
````

## üìå **Now for the serious plots...** üé®  
Let's wrap them in a **single function** that we can call repeatedly to generate the plots efficiently.

By doing this, we avoid repeating code and make our plotting process smoother and more modular.

````python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
````

````python
def multiplePlots(series):

    fig, axs = plt.subplots(2, 2)
    plt.tight_layout(pad=0.4, w_pad=4, h_pad=1.0)

    # Histogram
    sns.histplot(series, ax=axs[0, 0])

    # Lag plot
    lag = series.copy()
    lag = np.array(lag[:-1])
    current = series[1:]
    ax = sns.regplot(x=current, y=lag, fit_reg=False, ax=axs[0, 1])
    ax.set_ylabel("y_i-1")
    ax.set_xlabel("y_i")

    # QQ plot (updated 2023.08.18.1.CT to match usual QQ plot -- axes flipped)
    qntls, xr = stats.probplot(series, fit=False)
    sns.regplot(y=xr, x=qntls, ax=axs[1, 0])

    # Run sequence
    ax = sns.regplot(x=np.arange(len(series)), y=series, ax=axs[1, 1])
    ax.set_ylabel("val")
    ax.set_xlabel("i")
````
